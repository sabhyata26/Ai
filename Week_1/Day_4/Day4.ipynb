# FILE HANDLING AND EXCEPTIONS

## OPENING AND READING A FILE

```python
f = open("test.txt","r")
print(f.read())
f.close()
```

```python
f = open("test.txt","r")
print(f.readline())
f.close()
```

```python
f = open("test.txt","r")
print(f.read(1))
f.close()
```

# WITH OPERATION

```python
f = open("test.txt","r")
print(f.read())
f.close()
```

```python
with open("test.txt","r") as f:
    print(f.read())
```

## WRITING TO A FILE

```python
f = open("test.txt","w")
f.write("I am writing to this file")
f.close()
```

```python
with open("test.txt","w") as f:
    f.write("I wrote the file through with")
```

## APPENDING TO A FILE

```python
f = open("test.txt","a")
f.write("\nI am appending to this file")
f.close()
```
