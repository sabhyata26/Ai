#1.MEASURE OF CENTRAL TENDANCY

Note!! We can use module Statistics instead of numpy and pandas for mathematical statistics in python

```python
import pandas as pd

data =pd.Series([34,54,50,64,111,98,50,50,50,64,64])
#using .mean() .median() .mode() function from pandas module
mean=data.mean()
print("The mean of the data is:",int(mean))
median=data.median()
print("The median of data is:",int(median))
#using [0] to remove indexing
mode=data.mode()[0]
print("The mode of data is:",mode)
```

#Variability

```python
#finding range from a series
data =pd.Series([34,54,50,64,111,98,50,50,50,64,64])
#using .max() and .min() function from pandas module
range=data.max()-data.min()
print("The range of data is:",range)
```

#Variance and SD

```python
data =pd.Series([34,54,50,64,111,98,50,50,50,64,64])
#.var() for variance
variance=data.var()
print("The variance of data is:",variance)
#.std() for standard deviation
sd=data.std()
print("The standard deviation of data is:",sd)
```
