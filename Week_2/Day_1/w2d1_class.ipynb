# 1. MEASURE OF CENTRAL TENDENCY

#### Note!!!! We can also use module Statistics instead of Numpy and Pandas for mathematical statistics in Python

```python
import pandas as pd

data = pd.Series([34,67,39,69,98,111])

mean = data.mean()
print("The mean of the data is:",mean)
```

```python
data = pd.Series([34,67,39,69,34,65,78,34,98,111])

mode = data.mode()[0]
print("The mode of the data is:",mode)
```

# 2. MEASURE OF VARIABILITY

## Exercise 4: Calculating Range

```python
data = pd.Series([34,67,39,69,34,65,78,34,98,111])

range = data.max() - data.min()
print("The range of the data is:",range)
```

# VARIANCE AND SD

```python
data = pd.Series([34,67,39,69,34,65,78,34,98,111])

variance = data.var()
standard_deviation = data.std()
print(variance)
print(standard_deviation)
```

## VISUALIZING DATA WITH MATPLOTLIB
